# Web Scraping NFT data 

Data collected from <a href="https://www.cryptoslam.io/">CryptoSlam</a>

***

```python
# Modules needed
import requests  
import pandas as pd
import pymysql
from sqlalchemy import create_engine
import MySQLdb 
```

```python
# List with the names of the collections
page1=requests.get("https://web-api.cryptoslam.io/web-api/v1/sales/rank-by-volume?timeRange=all&orderBy=rank&sortDirection=asc&offset=0&limit=1000")
data=page1.json()['saleSummaries']
collections=[]
i=0
while i<1000: # Top 1000 collections
    collections.append(data[i]['productPath'])
    i+=1
```

```python
# Connecting to the database
db_name='nft'
db_host='localhost'
db_usarname='root'
db_password='mysqlpassword'
try:
    conn=pymysql.connect(host=db_host,port=int(3306),user='root',passwd=db_password,db=db_name)
    print('Connected to database:',db_name)
except e:
    print(e)
```

```python
engine = create_engine('mysql://root:mysqlpassword@localhost:3306/nft')
for collection in collections:
    page = requests.get("https://web-api.cryptoslam.io/web-api/v1/sales/"+collection+"/summary?groupingtype=daily")
    data=page.json()
    try: 
        df=pd.json_normalize(data['salesSummaryRecords'])
        df.to_sql(data['collectionName'], engine, if_exists='replace') # Write records stored in a DataFrame to a SQL database
    except:
        pass # In case there is information missing at some endpoint we just skip that collection data
```
